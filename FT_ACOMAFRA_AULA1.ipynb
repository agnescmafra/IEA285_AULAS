Usaremos o recurso de biblioteca [pint](https://pint.readthedocs.io/en/latest/) para incorporar o recurso de transformação de unidades no Python.

```python
from pint import UnitRegistry
ur = UnitRegistry()
```

**Exercício Resolvido 1:**
---

Escreva o peso molecular do CaCO3.


```python
# Sabe-se que :  C = 12g/mol ; Ca = 40,078g/mol ; O = 16g/mol

# Peso molecular do CaCO3 = C+Ca+3*O

C = 12*ur.grams/ur.mol
Ca = 40.078*ur.grams/ur.mol
O = 16*ur.grams/ur.mol

Peso = C+Ca+3*O

print(Peso)
```

**Exercício Resolvido 2:**
---

Transforme o Peso molecular do CaCO3 calculado no exercício 2 em Kg/milimol.

```python
# Convertendo a unidade com a ferramenta 
# OBJETO.ito(ur.UNIDADEDESEJADA)
# print(OBJETO)

Peso.ito(ur.kilogram/ur.millimol)
print(Peso)
```

**Exercício Dirigido 1:**
---

Converta usando a ferramenta ito() em Python:

a) 32.2 ft/s² para cm/ano²

b) 15 N/cm² para bar

**Resolução:**
---

```python
#Resolução a)
a = 32.2*ur.ft/ur.sec/ur.sec

a.ito(ur.centimeter/ur.year/ur.year)

print(a)
```

```python
#Resolução b)
b = 15*ur.N/ur.centimeter/ur.centimeter

b.ito(ur.bar)

print(b)
```
